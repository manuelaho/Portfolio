# Aufgaben zu BeautifulSoup

Die Webseite *https://hckrnews.com/*   fasst aktuelle Technologienachrichten in einer Liste zusammen. Wir wollen mit `BeautifulSoup` diese Informationen auslesen.

## Aufgabe 1:  Daten extrahieren


- Lass die URL *https://hckrnews.com/* über `requests` auslesen. Betrachte das Ergebnis mit BeautifulSoup. 

- Versuche herauszufinden, unter welchem Tag die Artikel dargestellt werden.

- Lass dir alle Überschriften der Artikel ausgeben. 

- Lass dir zusätzlich noch den Link ausgeben, womit man zum Artikel kommt.

```python
import requests 
from bs4 import BeautifulSoup 
```

```python
req = requests.get("https://hckrnews.com/")
```

```python
soup = BeautifulSoup(req.text, "html.parser")
soup.prettify
```

```python
# Tags für die Artikel
for i in soup.find_all("span",{"class":"source"}):
    print(i.text)
```

```python
# Überschriften Artikel
for i in soup.find_all("a",{"class":"link span15 story"}):
    print(i.text)
```

```python
# Links zum Artikel ausgeben 
for i in soup.find_all("a",{"class":"link span15 story"}):
    print(i.get("href"))
```

## Aufgabe 2: Filter einbauen

- Baue einen Filter deiner Wahl ein. Lasse dir z.b. nur die Artikel anzeigen, die das Wort `Python` enthalten.

```python
for i in soup.find_all("a",{"class":"link span15 story"}):
    if "Python" in i.text or  "Apple" in i.text:
        print(i.text)
```

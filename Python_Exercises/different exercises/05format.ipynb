## Aufgabe 1: Ausgabe von Variablen

- Lass den Nutzer zwei Zahlen eingeben und speichere diese Werte in den Variablen `x` und `y`. 
- Gib das Produkt der Zahlen auf der Konsole aus, *ohne* eine weitere Variable zu definieren.

**Beispiel:**
> Zahl x : 2<br>
Zahl y : 5<br>
2 mal 5 macht 10!

```python
x = input("Gib eine Zahl ein: ")
y = input("Gib eine weitere Zahl ein: ")

print(f"{x} mal {y} macht {int(x) * int(y)}!")
```

## Aufgabe 2: Nachkommastellen

- Lass den Nutzer zwei Zahlen eingeben und speichere diese Werte in den Variablen `x` und `y`. 
- Gib die Division der Zahlen mit nur einer Nachkommastelle auf der Konsole aus, _ohne_ eine weitere Variable zu definieren.

**Beispiel:**

> Zahl x : 5<br>
Zahl y : 15<br>
5 durch 15 macht 0.3!

```python
x = input("Gib eine Zahl ein: ")
y = input("Gib eine weitere Zahl ein: ")

print(f"{x} durch {y} macht {int(x) / int(y):.1f}!")
```

## Aufgabe 3: Tausendertrennzeichen

- Lass den Nutzer eine große Zahl eingeben und gib den Wert mit Unterstrichen zwischen den Tausenderstellen auf der Konsole aus.
> Zahl x : 1412535265423<br>
Du hast 1_412_535_265_423 angegeben.


```python
user = input("Gib eine Zahl mit 10 Stellen ein: ")
print(f"Du hast {int(user):_} angegeben!")
```

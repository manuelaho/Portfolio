## 2. Aufgabe: Manager & Mitarbeiter

- Wie viele Manager gibt es? Finde dazu die Anzahl von `Führungskräften` in der Spalte `Beschaeftigung`. 
- Stelle die Anzahl an `Führungskräften` und `Mitarbeitern` gegenüber.

```python
# Wie viele Manager gibt es? Finde dazu die Anzahl von Führungskräften in der Spalte Beschaeftigung

df[df["Beschaeftigung"] == "Führungskraft"].Beschaeftigung.count()
```

```python
# Stelle die Anzahl an Führungskräften und Mitarbeitern gegenüber

df[df["Beschaeftigung"] == "Mitarbeiter"].Beschaeftigung.count() 

führung = df[df["Beschaeftigung"] == "Führungskraft"].Beschaeftigung.count()
print(f"Führungskräfte: {führung:^30}")

mitarbeiter = df[df["Beschaeftigung"] == "Mitarbeiter"].Beschaeftigung.count() 
print(f"Mitarbeiter: {mitarbeiter:^30}")
```

## 3. Aufgabe: Männer & Frauen

- Stelle das Gehalt der Frauen und Männer gegenüber, z.B. durch ein **Histogramm**. Was kann man im Schaubild erkennen? 

> <span style="color:blue"> **_INFO:_** </span> Der Befehl [`hist`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.hist.html) könnte hier helfen. Besonders der Parameter `by`.

- Was verdienen Männer im Schnitt? Was Frauen?

```python
# Stelle das Gehalt der Frauen und Männer gegenüber

df[["Mitarbeiter_Gehalt", "Geschlecht_Mitarbeiter"]].hist(by = "Geschlecht_Mitarbeiter")
```

```python
# Was verdienen Männer im Schnitt? Was Frauen?

df.groupby(["Geschlecht_Mitarbeiter"]).Mitarbeiter_Gehalt.mean()
```

## 4. Aufgabe: Gehaltserhöhung

- Die Spalte `Gehaltserhoehung` gibt an, ob die Person eine Gehaltserhöhung bekommt. Sollte hier "JA" stehen, berechne einen Aufschlag von 10% auf das aktuelle Gehalt. 

> <span style="color:blue"> **_INFO:_** </span> Der Befehl [`apply`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.apply.html) könnte hier helfen. Damit kann man eine Funktion auf _jeden_ Eintrag einer Spalte anwenden.

```python
def gehaltserhöhung(x):
        return x * 1.1 
    
df[df["Gehaltserhoehung"] == "JA"].Mitarbeiter_Gehalt.apply(gehaltserhöhung)
```

## 5. Aufgabe: Alter & Beschäftigung

- Wie ist der Name des jüngsten Mitarbeiters des Unternehmens?

- Wer arbeitet am schon am längsten im Unternehmen?



```python
# Wie ist der Name des jüngsten Mitarbeiters des Unternehmens?

geburtstag = df[df["Bday_Mitarbeiter"] == df["Bday_Mitarbeiter"].max()]

geburtstag[["MA_Nachname", "MA_Vorname","Bday_Mitarbeiter" ]].sort_values(by = "MA_Nachname")
```

```python
# Wer arbeitet schon am längsten im Unternehmen?

längsten = df[df["Hire_Date_Mitarbeiter"] == df["Hire_Date_Mitarbeiter"].min()]

längsten[["MA_Nachname", "MA_Vorname","Hire_Date_Mitarbeiter" ]].sort_values(by = "MA_Nachname")
```

# alle features:

```python
X01 = mitarbeiter02.drop(['Attrition'], axis=1)
y01 = mitarbeiter02['Attrition']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X01, y01, test_size = 0.20, random_state = 33)
```

```python
# Standardscaler fÃ¼r Streuung und Varianz

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
# Algorithmen
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier

log = LogisticRegression()
knc = KNeighborsClassifier()
svc = SVC()
nab = GaussianNB()
rfc = RandomForestClassifier()
```

```python
# Ellbow method fÃ¼r KN um die geeignete Anzahl von Nachbarn auszuwÃ¤hlen --> jedes k durchprobieren

fehler = []

for k in range(1, 10):                                                    
    knc = KNeighborsClassifier(n_neighbors = k)        
    knc.fit(X_train, y_train)
    predictions_k = knc.predict(X_test)
    fehler.append(np.mean(predictions_k != y_test))     

plt.figure(figsize=(6, 4))
plt.plot(range(1, 10), fehler, marker = "x")     
plt.xlabel('n_neighbors')
plt.ylabel('Fehlerwert')
plt.title("Ellbow method 01")
#plt.savefig(r'C:\Users\Fujitsu\Documents\Data Science_Manu\Projekt\Human_ressource10.png')
```

```python
# beste Parameter fÃ¼r SVC

from sklearn.model_selection import GridSearchCV

# dictionary mit Werten fÃ¼r C und gamma
hyperparameter = {"C" : [0.1, 1, 10, 100, 1000], "gamma" : [1, 0.1, 0.01, 0.001, 0.0001]}

grid01 = GridSearchCV(SVC(), hyperparameter, refit = True)
grid01.fit(X_train, y_train)
```

```python
grid01.best_params_
```

```python
# TRAINIEREN

log = LogisticRegression().fit(X_train, y_train)
knc = KNeighborsClassifier(n_neighbors = 6).fit(X_train, y_train)              # bester Wert 6
svc = SVC(C=1000, gamma=0.0001).fit(X_train, y_train)
nab = GaussianNB().fit(X_train, y_train)
rfc = RandomForestClassifier(random_state = 33, n_estimators = 1000).fit(X_train, y_train)
```

```python
# VORHERSAGEN

pred_log = log.predict(X_test)
pred_knc = knc.predict(X_test)
pred_svc = svc.predict(X_test)
pred_nab = nab.predict(X_test)
pred_rfc = rfc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, mean_absolute_error, confusion_matrix, classification_report
```

```python
# REPORTS

print("Genauigkeit LogReg: {:.2f}%".format((accuracy_score(y_test, pred_log)*100)))
print("Genauigkeit KNN: {:.2f}%".format((accuracy_score(y_test, pred_knc)*100)))        # verschiedene k ausprobiert  
print("Genauigkeit SVC: {:.2f}%".format((accuracy_score(y_test, pred_svc)*100)))                      
print("Genauigkeit NAIVE: {:.2f}%".format((accuracy_score(y_test, pred_nab)*100)))
print("Genauigkeit Random: {:.2f}%".format((accuracy_score(y_test, pred_rfc)*100)))            
```

```python
models = [log, knc, svc, nab, rfc]
model_names = ["LogisticRegression","KNeighborsClassifier", "SVC", "NaiveBayes","RandomForestClassifier",]


# for- Schleife bauen um Ã¼ber alle Algorithmen zu iterieren mit dazugehÃ¶rigen reports

accuracy_scores = []
precision_scores = []
recall_scores = []
error_rate = []

for i, m in enumerate(models):
    y_pred = m.predict(X_test)                               # Vorhersagen fÃ¼r alle Algorithmen
    conf_matrix = confusion_matrix(y_test, y_pred)           # confusion matrix
    
    print("Modell: ", model_names[i], "\n")
    print("Classification Report: ", classification_report(y_test, y_pred), "\n")
    print("Confusion Matrix: \n", confusion_matrix(y_test, y_pred), "\n")
    
    # nur fÃ¼r Teil um confusion matrix in _Tabelle zu erstellen mit Dataframe
    true_positive = conf_matrix[0,0]
    true_negative = conf_matrix[1,1]
    false_positive = conf_matrix[0,1]
    false_negative =  conf_matrix[1,0]
    
    #TP: Als Klasse 0 vorhergesagt, tatsÃ¤chlich Klasse 0
    #TN: Als Klasse 1 vorhergesagt, tatsÃ¤chlich Klasse 1
    #FP: Als Klasse 0 vorhergesagt, tatsÃ¤chlich Klasse 1
    #FN: Als Klasse 1 vorhergeasgt, tatsÃ¤chlich Klasse 0
    
    total = true_positive + true_negative + false_positive + false_negative  
    
    accuracy = (true_positive + true_negative) / total                 # accuracy berechnen usw.....
    precision = true_positive / (true_positive + false_positive)
    recall = true_positive / (true_positive + false_negative)
    error = (false_positive + false_negative) / total
    
    
    accuracy_scores.append(accuracy.round(2)*100)
    precision_scores.append(precision.round(2))
    recall_scores.append(recall.round(2))
    error_rate.append(error.round(2))

# Dataframe erstellen:

ergebnisse_df = pd.DataFrame(list(zip(accuracy_scores, precision_scores, recall_scores, error_rate)), 
                            index = model_names, columns = ["accuracy", "precision", "recall", "error"])

ergebnisse_df
```
